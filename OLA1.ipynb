```python
import numpy as np
import sympy as sp
sp.init_printing(use_latex=True)
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(-5, 5)
y = 1/(1 + np.exp(-x))
plt.plot(x, y)
plt.title(r'$y = \frac{1}{1 + e^{-x}}$')
```

### Problem 3

```python
x = sp.symbols('x')
y = 1/(1 + sp.exp(-x))
dydx = y.diff(x)
dydx
```

```python
x_vals = np.linspace(-5, 5)
y_vals = np.array([dydx.evalf(subs=dict(x=x_val)) for x_val in x_vals])
plt.plot(x_vals, y_vals)
plt.title(r'$y = %s$'%(sp.latex(dydx)))
```
